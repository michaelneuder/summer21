```python
import numpy as np
```

```python
def haydens(a,b):
    return 1-np.power(a/b,1/4)

def getX(k, p):
    return np.sqrt(k/p)

def getY(k, p): 
    return np.sqrt(k*p)

def mine(a, b, r_a, r_b):
    k = r_a * r_b
    c = np.sqrt(a*b)
    
    x_c = getX(k, c)
    x_b = getX(k, b)
    x_r = (x_c - x_b) / r_a
    
    y_c = getY(k, c)
    y_a = getY(k, a)
    y_r = (y_c - y_a) / r_b
    
    return np.sqrt(x_r * y_r)
```
